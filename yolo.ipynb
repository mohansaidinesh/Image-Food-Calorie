```python
from ultralytics import YOLO
import torch

# Check GPU
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print("Using device:", device)

# Load YOLOv8 Classification Model
model = YOLO("yolov8m-cls.pt")   # options: yolov8n-cls, yolov8s-cls, yolov8m-cls

# Train the model
model.train(
    data="Indian Food Images\Indian Food Images",
    epochs=15,
    imgsz=200,
    batch=32,
    optimizer="Adam",
    lr0=1e-4,
    device=device,
    project="IndianFood_YOLOv8",
    name="YOLOv8_Classification"
)
```

```python
metrics = model.val()
print(metrics)
```

```python
results = model.predict(
    source="test_image.jpg",
    imgsz=200,
    device=device
)

# Print class & confidence
for r in results:
    print(r.probs.top1, r.names[r.probs.top1], r.probs.top1conf)
```
